````python
count = 0.00
num = 0
ix = 0
for i in df['gdpPercap']:
    num = num + 1
    if count > i:
        count = count
    else:
        count = i
        ix = num
df.query('index == @ix-1')
````

# <추가문제>

**Exercise 8**: Who produces more good rated movies, Steven Spielberg or James Cameron? Select the `director_name` column, but only  when said `director_name` is one of those two options, _and_ the `imdb_score` column is greater than or equal to 8.0.

````python
movie = pd.read_csv('movie.csv')
````

````python
movie_1 = movie.loc[((movie['director_name']=='Steven Spielberg') | (movie['director_name']=='James Cameron')) &
           (movie['imdb_score'] >= 8.0),]
count = movie_1['director_name'].value_counts()
if count['Steven Spielberg'] > count['James Cameron']:
    print('Steven Spielberg')
else:
    print('James Cameron')
````
